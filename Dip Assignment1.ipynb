**Part 2**

gamma_transformation

```python
def gamma_transformation(img, gamma):
    normalize_img = img/255.0
    gamma_corrected= np.power(normalize_img, gamma)
    return gamma_corrected

gamma_img1= gamma_transformation(img,1)
gamma_img2= gamma_transformation(img,3)
gamma_img3= gamma_transformation(img,0.4)
plt.figure(figsize=(15, 8))

    # Original image
plt.subplot(1, 4, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

    # Gamma = 1
plt.subplot(1, 4, 2)
plt.imshow(gamma_img1)
plt.title('Gamma = 1')
plt.axis('off')

    # Gamma = 3
plt.subplot(1, 4, 3)
plt.imshow(gamma_img2)
plt.title('Gamma = 3')
plt.axis('off')

    # Gamma = 0.4
plt.subplot(1, 4, 4)
plt.imshow(gamma_img3)
plt.title('Gamma = 0.4')
plt.axis('off')
plt.show()
```

**Part 3**


logritimic_transofrmation

```python
def logritimic_transofrmation(img,c):
    normalize_img = img/255.0
    log_img = c*np.log(1+normalize_img)
    return log_img

Log_img1= logritimic_transofrmation(img,1)
Log_img2= logritimic_transofrmation(img,2)
Log_img3= logritimic_transofrmation(img,5)

plt.figure(figsize=(15, 8))

plt.subplot(1, 4, 1)
plt.imshow(img)
plt.title("Orignal Image")
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(Log_img1)
plt.title("c=1")
plt.axis('off')


plt.subplot(1,4,3)
plt.imshow(Log_img2)
plt.title("c=2")
plt.axis('off')


plt.subplot(1,4,4)
plt.imshow(Log_img3)
plt.title("c=5")
plt.axis('off')
plt.show()
```

**Part 4**

Image Contrast Streaching

```python
def contrast_streaching(img,m):
  normalize_img = img/255.0
  contrased_streached = 1 / (1+(m/(normalize_img + 1e-5))**2)
  return contrased_streached

streached_img1 = contrast_streaching(img,0.2)
streached_img2 = contrast_streaching(img,0.3)
streached_img3 = contrast_streaching(img,0.7)

plt.figure(figsize = (15 , 8))

plt.subplot(1,4,1)
plt.imshow(img)
plt.title("Orignal Image")
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(streached_img1)
plt.title("m = 0.2")
plt.axis('off')

plt.subplot(1,4,3)
plt.imshow(streached_img2)
plt.title("m = 0.3")
plt.axis('off')

plt.subplot(1,4,4)
plt.imshow(streached_img3)
plt.title("m = 0.7")
plt.axis('off')
plt.show()
```
